```python
import os
import numpy as np
import pandas as pd
import cv2

filepath = '../glaucoma_fundus.csv'
df = pd.read_csv(filepath)

# a dict with the image as key and dx as value
image_dx_dict = {}
for i in range(len(df['image'])):
    image_dx_dict[df['image'][i]] = df['dx'][i]

# first two '-' as patient name, value as imagename, to a dictionary
patientid_name_dict = {}
for i in range(len(df['image'])):
    if '-' in df['image'][i]:
        patientid = df['image'][i].split('-')[0] + '-' + df['image'][i].split('-')[1]
        imagename = df['image'][i]
        if patientid not in patientid_name_dict:
            patientid_name_dict[patientid] = [imagename]
        else:
            patientid_name_dict[patientid].append(imagename)
    
# random 70% dictionary name as train, 30% as test
train_images = []
test_images = []
for key in patientid_name_dict:
    if np.random.rand() < 0.75:
        train_images += patientid_name_dict[key]
    else:
        test_images += patientid_name_dict[key]

# delete folder train and test if exists
if os.path.exists('./train'):
    os.system('rm -rf ./train')
if os.path.exists('./test'):
    os.system('rm -rf ./test')

# create folder train, and 0 and 1 in train
if not os.path.exists('./train'):
    os.mkdir('./train')
    os.mkdir('./train/0')
    os.mkdir('./train/1')

# create folder test, and 0 and 1 in test
if not os.path.exists('./test'):
    os.mkdir('./test')
    os.mkdir('./test/0')
    os.mkdir('./test/1')

imagefolder = 'D:\\KEN\\'

# copy images to train/0, train/1, test/0, test/1
import shutil
for image in train_images:
    # shutil.copy(imagefolder + image, './train/' + str(image_dx_dict[image]) + '/')
    try:
        img = cv2.imread(imagefolder + image)
        # resize to half
        img = cv2.resize(img, (0,0), fx=0.5, fy=0.5)
        cv2.imwrite('./train/' + str(image_dx_dict[image]) + '/' + image, img)
    except Exception as e:
        print(image, e)

for image in test_images:
    try:
        img = cv2.imread(imagefolder + image)
        # resize to half
        img = cv2.resize(img, (0,0), fx=0.5, fy=0.5)
        cv2.imwrite('./test/' + str(image_dx_dict[image]) + '/' + image, img)
    except Exception as e:
        print(image, e)
```
